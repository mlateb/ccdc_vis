```python
%matplotlib inline
from sklearn.neighbors import KernelDensity
from ipywidgets import StaticInteract, RangeWidget, RadioWidget, interact
import numpy as np
import os, rasterio
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from osgeo import gdal, gdal_array
import Image
from IPython.html import widgets
from IPython.display import display
```

This is an example of a potential visualation tool for the output of Continuous Change Detection and Classification (CCDC) analysis.

```python
#Get list of files
imagelist=[]
for subdir, dirs, files in os.walk('./', followlinks=True):
    for file in files:
        filepath = subdir + os.sep + file
        if filepath.endswith(".tif"):
            imagelist.append(filepath)
imnum=len(imagelist)

#Get image attributes
with rasterio.open(imagelist[0], 'r+') as r:
    arr = r.read()
    shape=arr.shape
    
#Initiate arrays to hold change date. Easiest way found to center color stretch. 
stack=np.zeros((imnum,shape[0],shape[1],shape[2]))
mstack=np.ma.zeros((imnum,shape[0],shape[1],shape[2]))

#Create image stack
count=0
for n in imagelist:
    with rasterio.open(n, 'r+') as r:
        stack[count,:,:,:] = r.read()
        stack[stack==0] = np.nan
        stack[count,0,0,0] = 2014
        mstack[count,:,:,:]=np.ma.masked_where(stack[count,:,:,:]==0,stack[count,:,:,:])
        count+=1
        
#Define class color (thanks Chris Holden: http://ceholden.github.io/open-geo-tutorial/python/chapter_5_classification.html)
n = cover.max()
colors = dict((
    (2, (0, 0, 100, 255)),  
    (4, (0, 50, 0, 255)), 
    (6, (0, 0, 100, 255)), 
    (7, (0, 128, 0, 255)), 
    (8, (160, 82, 45, 255)),  
    (9, (25, 25, 112, 255))  
))
for k, v in colors.iteritems():
    _v = [_v / 255.0 for _v in v]
    colors[k] = _v 
index_colors = [colors[key] if key in colors else 
                (0, 0, .5, 0) for key in xrange(1, n + 1)]
cmap = plt.matplotlib.colors.ListedColormap(index_colors, 'Classification', n)

#Define function to create map
def sigmoid_demo(a=1984, b='Change'):
    if b=='Change':
        band=-1984+a
        year=mstack[band,0,:,:]
        plt.figure(figsize=(5,5))
        plt.subplot(122)
        ax1 = plt.axes(frameon=False)
        ax1.axes.get_yaxis().set_visible(False)
        ax1.get_xaxis().tick_bottom()
        ax1.axes.get_xaxis().set_visible(False)
        plt.imshow(year, cmap=plt.cm.YlOrRd)
        plt.title('Change Until %s' % a) 
    elif b=='Class':
        band=-1983+a
        fname = 'CoverMap_subset'
        cover_ds = gdal.Open(fname, gdal.GA_ReadOnly)
        cover = cover_ds.GetRasterBand(band).ReadAsArray().astype(np.uint8)
        plt.figure(figsize=(5,5))
        plt.subplot(122)
        ax1 = plt.axes(frameon=False)
        ax1.axes.get_yaxis().set_visible(False)
        ax1.get_xaxis().tick_bottom()
        ax1.axes.get_xaxis().set_visible(False)
        plt.imshow(cover, cmap=cmap, interpolation='none')
        plt.title('Landcover %s' % a) 
#Create slider for years
a_slider = widgets.IntSliderWidget(min=1984, max=2013, step=1, value=1984)

#Create dropdown menu and display
drop = widgets.DropdownWidget(values={'Class': 'Class','Change': 'Change'})
w=widgets.interactive(sigmoid_demo,a=a_slider, b=drop)
display(w)
print "Control slider to change year of map and use dropdown menu to select map type."
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plot(amplitude, color):
    fig, ax = plt.subplots(figsize=(4, 3),
                           subplot_kw={'axisbg':'#EEEEEE',
                                       'axisbelow':True})
    ax.grid(color='w', linewidth=2, linestyle='solid')
    x = np.linspace(0, 10, 1000)
    ax.plot(x, amplitude * np.sin(x), color=color,
            lw=5, alpha=0.4)
    ax.set_xlim(0, 10)
    ax.set_ylim(-1.1, 1.1)
    return fig

from ipywidgets import StaticInteract, RangeWidget, RadioWidget

StaticInteract(plot,
               amplitude=RangeWidget(0.1, 1.0, 0.1),
               color=RadioWidget(['blue', 'green', 'red']))
```
